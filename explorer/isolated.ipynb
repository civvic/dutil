**The Tricky Part: Making It Runnable**
If you extract just a function, you likely lose its imports and dependencies. Two approaches:

- **Dependency tracing**: Analyze the target messages, find what names they use, trace back to which messages define/import those names, auto-include them
- **Post-hoc validation**: Create the isolated dialog, run `check_flakes` for `UndefinedName`/`UnusedImport`, then either auto-fix or report to user

I'd lean toward a hybrid: do basic import tracing automatically (it's usually obvious which import messages to include), then validate with `check_flakes` and surface any remaining issues.
