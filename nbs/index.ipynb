## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/civvic/dutil.git
```

or from [conda][conda]

```sh
$ conda install -c civvic dutil
```

or from [pypi][pypi]


```sh
$ pip install dutil
```
